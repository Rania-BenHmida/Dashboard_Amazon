```python
import os
import pandas as pd

# Path to the folder
folder_path = 'raw data'

# Dictionary to store named DataFrames
named_dataframes = {}

# List to store names of the DataFrames
csv_names = []

# Read and name each file
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        name = filename.replace('.csv', '').replace(' ', '_')  # Clean name
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        named_dataframes[name] = df  # Store in dictionary
        csv_names.append(name)

print(f"Loaded {len(csv_names)} CSV files.")
print("Names:", csv_names)
```

```python
# Example: Access the DataFrame for "All Appliances.csv"
df_appliances = named_dataframes['Travel_Duffles']
print(df_appliances.head())
```

```python
for i in csv_names:
 # named_dataframes[i].info()
  print(named_dataframes[i].shape)

```

```python
for name in csv_names:
    df = named_dataframes[name]  # Access the DataFrame

    # Clean and convert 'no_of_ratings'
    if 'no_of_ratings' in df.columns:
        df['no_of_ratings'] = df['no_of_ratings'].astype(str).str.replace(r',', '', regex=True)
        df['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'], errors='coerce')

    # Convert 'ratings' to numeric
    if 'ratings' in df.columns:
        df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

    # Clean and convert price columns
    if 'discount_price' in df.columns:
        df['discount_price'] = df['discount_price'].astype(str).str.replace(r'[₹, ]', '', regex=True).astype(float)
    if 'actual_price' in df.columns:
        df['actual_price'] = df['actual_price'].astype(str).str.replace(r'[₹, ]', '', regex=True).astype(float)

    # Remove (Renewed) from name column
    if 'name' in df.columns:
        df['name'] = df['name'].astype(str).str.replace(r'\(Renewed\)', '', regex=True)
    df['brand'] = df['name'].str.split(' ').str[0]

    # Show null value counts per file
    print(f"\n---- {name} ----")
    print(df.isnull().sum())

    # Save changes back to the dictionary
    named_dataframes[name] = df
```

```python
import os
import pandas as pd

# Step 1: Define category groupings
category_groups = {
    'Home_Products': ['home & kitchen', 'home, kitchen, pets', 'appliances'],
    'Electronics_and_Media': ['tv, audio & cameras', 'music'],
    'Fashion': ["women's clothing", "men's clothing", "women's shoes", "men's shoes", "kids' fashion", 'accessories'],
    'Groceries_and_Retail': ['grocery & gourmet foods', 'stores', 'pet supplies'],
    'Lifestyle': ['sports & fitness', 'bags & luggage', 'beauty & health'],
    'Automotive_and_Tools': ['car & motorbike', 'industrial supplies'],
    'Kids_and_Toys': ['toys & baby products']
}

# Step 2: Initialize containers
folder_path = 'raw data'
named_dataframes = {}
csv_names = []
grouped_files = {group: [] for group in category_groups}

# Step 3: Read all CSV files
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        name = filename.replace('.csv', '').replace(' ', '_')
        file_path = os.path.join(folder_path, filename)
        try:
            df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
            named_dataframes[name] = df
            csv_names.append(name)

            # Step 4: Determine group by main_category
            if 'main_category' in df.columns:
                main_cats = df['main_category'].dropna().str.lower().unique()
                for group, valid_values in category_groups.items():
                    if any(cat in valid_values for cat in main_cats):
                        grouped_files[group].append(name)
                        break  # only assign to one group

        except Exception as e:
            print(f"❌ Failed to read {filename}: {e}")

# Step 5: Display the result
print("\n📊 Grouped Files by Category:\n")
for group, files in grouped_files.items():
    print(f"🔹 {group} ({len(files)} files):")
    for f in files:
        print(f"  - {f}")
    print()
```

```python
import pandas as pd

# Read the CSV
df4 = pd.read_csv('amazon - untouched.csv',  on_bad_lines='skip')


# Display basic structure
print("\n🧾 DataFrame Info:")
print(df4.info())

# Display first few rows
print("\n👀 Preview of Data:")
print(df4.head())

# Check for null values
print("\n🔍 Null Values Per Column:")
print(df4.isnull().sum())

# Show unique values per column (optional, useful for columns like category or brand)
print("\n📊 Unique Values Example:")
for col in df4.columns:
    print(f"{col}: {df4[col].nunique()} unique values")
```

```python
# Afficher les valeurs uniques de la colonne 'category'
print(df4['category'].unique())
```

```python
 #Function to extract main and sub category
def extract_categories(value):
    if pd.isna(value):
        return pd.Series([None, None])
    parts = value.split('|')
    if value.startswith('Computers&Accessories'):
        return pd.Series(['Electronics', 'Computers&Accessories'])
    elif len(parts) >= 2:
        return pd.Series([parts[0], parts[1]])
    else:
        return pd.Series([parts[0], None])  # handle cases with no sub-category

# Apply the function
df4[['main_category', 'sub_category']] = df4['category'].apply(extract_categories)

# Show unique values of each
print("✅ Unique main_category values:")
print(df4['main_category'].unique())

print("\n✅ Unique sub_category values:")
print(df4['sub_category'].unique())
```

```python
df4.head()
```

```python
# Drop the original 'category' column
df4 = df4.drop(columns=['category'])

# Optional: Confirm it's gone
print(df4.columns)
```

```python
# Rename and select desired columns
df4_cleaned = df4.rename(columns={
    'product_name': 'name',
    'img_link': 'image',
    'product_link': 'link',
    'rating': 'ratings',
    'rating_count': 'no_of_ratings',
    'discounted_price': 'discount_price'
})[[
    'name',
    'main_category',
    'sub_category',
    'image',
    'link',
    'ratings',
    'no_of_ratings',
    'discount_price',
    'actual_price'
]]
```

```python
# Clean values
df4_cleaned['ratings'] = pd.to_numeric(df4_cleaned['ratings'], errors='coerce')
df4_cleaned['no_of_ratings'] = pd.to_numeric(
    df4_cleaned['no_of_ratings'].astype(str).str.replace(',', ''), errors='coerce')

df4_cleaned['discount_price'] = df4_cleaned['discount_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df4_cleaned['actual_price'] = df4_cleaned['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)

# Extract brand from product name
df4_cleaned['brand'] = df4_cleaned['name'].astype(str).str.split().str[0]

# Save to new CSV
df4_cleaned.to_csv('amazon_cleaned.csv', sep=';', index=False)

print("✅ File 'amazon_cleaned.csv' created successfully.")
```

```python
import pandas as pd

# Lire le fichier CSV (avec le séparateur correct si nécessaire, ici on suppose ';')
df5 = pd.read_csv('electronics_data - edited.csv', sep=';')

# Afficher les informations générales
print("📊 Aperçu général :")
print(df5.info())

# Afficher les premières lignes
print("\n🔍 Premières lignes :")
print(df5.head())

# Vérifier les valeurs manquantes
print("\n🧼 Valeurs manquantes par colonne :")
print(df5.isnull().sum())
```

```python
df5.head()
```

```python
# Clean 'reduction_percentage' by removing '-' and '%'
df5['reduction_percentage'] = (
    df5['reduction_percentage']
    .astype(str)
    .str.replace('-', '', regex=False)
    .str.replace('%', '', regex=False)
)

# Convert to numeric (optional)
df5['reduction_percentage'] = pd.to_numeric(df5['reduction_percentage'], errors='coerce')

# Preview cleaned column
print(df5['reduction_percentage'].head())
```

```python
df5.drop(columns=['review_stars'], inplace=True)
df5.drop(columns=['last_month_sales'], inplace=True)

df5.head()
```

```python
# Remove 'ratings' and commas from 'number_of_reviews' and convert to numeric
df5['number_of_reviews'] = df5['number_of_reviews'].str.replace('ratings', '', regex=False)
df5['number_of_reviews'] = df5['number_of_reviews'].str.replace(',', '', regex=False)
df5['number_of_reviews'] = pd.to_numeric(df5['number_of_reviews'], errors='coerce')

# Remove 'out of 5' from 'average_review' and convert to float
df5['average_review'] = df5['average_review'].str.replace('out of 5', '', regex=False)
df5['average_review'] = pd.to_numeric(df5['average_review'], errors='coerce')
```

```python
df5.head()
```

```python
# Fill NaN values in 'brand' with the first word from the 'name' column
df5['brand'] = df5['brand'].fillna(df5['name'].str.split().str[0])
```

```python
df5.head()
```

```python
import pandas as pd

# Ensure reduction_percentage is float
df5['reduction_percentage'] = pd.to_numeric(df5['reduction_percentage'], errors='coerce')

# Ensure price is float
df5['price'] = pd.to_numeric(df5['price'], errors='coerce')

# Compute old_price (before discount) and discounted_amount
df5['old_price'] = df5['price'] / (1 - df5['reduction_percentage'] / 100)
df5['discounted_amount'] = df5['old_price'] - df5['price']

# Round for clarity
df5['old_price'] = df5['old_price'].round(2)
df5['discounted_amount'] = df5['discounted_amount'].round(2)

# Optional: view a sample
print(df5[['price', 'reduction_percentage', 'old_price', 'discounted_amount']].head())
```
